```python
import numpy as np
import math
from numpy import linalg as LA
import matplotlib.pyplot as plt

k=float(input())
m=1

mt=np.array([[2,-1,0],[-1/3,2/3,-1/3],[0,-1/2,1]])
ww,y=LA.eigh(mt)
w=np.matrix(np.zeros([3]))
y1=np.matrix(np.zeros([3]))
y2=np.matrix(np.zeros([3]))
y3=np.matrix(np.zeros([3]))
for i in range(3):
  w[0,i]=math.sqrt(ww[i]*k/m)
  y1[0,i]=y[i,0]
  y2[0,i]=y[i,1]
  y3[0,i]=y[i,2]
  #print("w",i+1,"=",w[0,i])
w=np.transpose(w)

t=np.arange(0,2*np.pi/w[0,0],0.01)
print("frequency = ",w[0,0], ", amplitudes = ", y1)
y1=np.transpose(y1)
for l in range(4):
  plt.plot(y1[0,0]*l*np.sin(w[0,0]*t), y1[0,0]*l*np.cos(w[0,0]*t),lw=1,color="mediumslateblue",)
  plt.plot(y1[1,0]*l*np.sin(w[0,0]*t), y1[1,0]*l*np.cos(w[0,0]*t),lw=1,color="skyblue",)
  plt.plot(y1[2,0]*l*np.sin(w[0,0]*t), y1[2,0]*l*np.cos(w[0,0]*t),lw=1,color="navy",)
plt.show()
```

```python
t=np.arange(0,2*np.pi/w[1,0],0.01)
print("frequency = ",w[1,0], ", amplitudes = ", y2)
y2=np.transpose(y2)
for l in range(4):
  plt.plot(y2[0,0]*l*np.sin(w[1,0]*t), y2[0,0]*l*np.cos(w[1,0]*t),lw=1,color="c")
  plt.plot(y2[1,0]*l*np.sin(w[1,0]*t), y2[1,0]*l*np.cos(w[1,0]*t),lw=1,color="darkgreen")
  plt.plot(y2[2,0]*l*np.sin(w[1,0]*t), y2[2,0]*l*np.cos(w[1,0]*t),lw=1,color="lightgreen")
plt.show()
```

```python
t=np.arange(0,2*np.pi/w[2,0],0.01)
print("frequency = ",w[2,0], ", amplitudes = ", y3)
y3=np.transpose(y3)
for l in range(4):
  plt.plot(y3[0,0]*l*np.sin(w[2,0]*t), y3[0,0]*l*np.cos(w[2,0]*t),lw=1,color="m")
  plt.plot(y3[1,0]*l*np.sin(w[2,0]*t), y3[1,0]*l*np.cos(w[2,0]*t),lw=1,color="plum")
  plt.plot(y3[2,0]*l*np.sin(w[2,0]*t), y3[2,0]*l*np.cos(w[2,0]*t),lw=1,color="mediumpurple")
plt.show()
```
