```python
import numpy as np
import math
from numpy import linalg as LA
import matplotlib.pyplot as plt
print("введите жесткость пружин")
k=float(input())
m=1
while k <= 0:
  print('Жесткость всегда положительна, пробуй ещё')
  k=float(input())

mt=np.array([[2,-1,0],[-1/3,2/3,-1/3],[0,-1/2,1]])
ww,y=LA.eigh(mt)
w=np.matrix(np.zeros([3]))
y1=np.matrix(np.zeros([3]))
y2=np.matrix(np.zeros([3]))
y3=np.matrix(np.zeros([3]))
for i in range(3):
  w[0,i]=math.sqrt(ww[i]*k/m)
  y1[0,i]=y[i,0]
  y2[0,i]=y[i,1]
  y3[0,i]=y[i,2]
  #print("w",i+1,"=",w[0,i])
w=np.transpose(w)

t=np.arange(0,2*np.pi/w[0,0],0.01)
print("frequency = ",w[0,0], ", amplitudes = ", y1)
fig, ax = plt.subplots()
y1=np.transpose(y1)
for l in range(4):
  ax.plot(y1[0,0]*l*np.sin(w[0,0]*t), y1[0,0]*l*np.cos(w[0,0]*t),lw=1,color="mediumslateblue",)
  ax.plot(y1[1,0]*l*np.sin(w[0,0]*t), y1[1,0]*l*np.cos(w[0,0]*t),lw=1,color="skyblue",)
  ax.plot(y1[2,0]*l*np.sin(w[0,0]*t), y1[2,0]*l*np.cos(w[0,0]*t),lw=1,color="navy",)
ax.set_xlabel('координата')
ax.set_ylabel('скорость')
ax.plot(0,0,lw=1,color = "mediumslateblue",label = "телега левая")
ax.plot(0,0,lw=1,color = "navy",label = "телега правая")
ax.plot(0,0,lw=1,color = "skyblue",label = "телега центр.")
ax.legend(loc='lower right')
plt.show()
```

```python
t=np.arange(0,2*np.pi/w[1,0],0.01)
fig, ax = plt.subplots()
print("frequency = ",w[1,0], ", amplitudes = ", y2)
y2=np.transpose(y2)
for l in range(4):
  ax.plot(y2[0,0]*l*np.sin(w[1,0]*t), y2[0,0]*l*np.cos(w[1,0]*t),lw=1,color="c")
  ax.plot(y2[1,0]*l*np.sin(w[1,0]*t), y2[1,0]*l*np.cos(w[1,0]*t),lw=1,color="darkgreen")
  ax.plot(y2[2,0]*l*np.sin(w[1,0]*t), y2[2,0]*l*np.cos(w[1,0]*t),lw=1,color="lightgreen")
ax.set_xlabel('координата')
ax.set_ylabel('скорость')
ax.plot(0,0,lw=1,color = "c",label = "телега левая")
ax.plot(0,0,lw=1,color = "lightgreen",label = "телега правая")
ax.plot(0,0,lw=1,color = "darkgreen",label = "телега центр.")
ax.legend(loc='lower right')
plt.show()
```
